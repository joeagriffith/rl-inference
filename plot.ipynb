```python
import matplotlib.pyplot as plt
import numpy as np
import json
data = []
for i in range(10):
    metrics = json.load(open(f"log_cup_catch_OG/{i}/metrics.json"))
    data.append(metrics["rewards"])
array = np.zeros((len(data), 49))
for i, d in enumerate(data):
    array[i, :len(d)] = d
array.shape
data = array.mean(axis=0)
plt.plot(data)
```
