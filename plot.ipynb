```python
data = []
for i in range (10):
    path = f'log_cup_catch_DQN-like_{i}/metrics.json'
    data.append(json.load(open(path)))
```

```python
rewards = np.zeros((10, len(data[0]['rewards'])))
for i in range(10):
    rewards[i] = data[i]['rewards']
rewards = np.mean(rewards, axis=0)
plt.plot(rewards, label='DQN-like')
plt.show()
```
