```python
import matplotlib.pyplot as plt
import numpy as np
import json
og_data = []
for i in range(10):
    metrics = json.load(open(f"log_cup_catch/OG/{i}/metrics.json"))
    og_data.append(metrics["rewards"])
dqn_data = []
for i in range(10):
    metrics = json.load(open(f"log_cup_catch/DQN-like/{i}/metrics.json"))
    dqn_data.append(metrics["rewards"])

array = np.zeros((len(og_data), 49))
for i, d in enumerate(og_data):
    array[i] = d
array.shape
og_data = array.mean(axis=0)
array = np.zeros((len(dqn_data), 49))
plt.plot(og_data, label="OG")
for i, d in enumerate(dqn_data):
    array[i] = d
array.shape
dqn_data = array.mean(axis=0)
plt.plot(dqn_data, label="DQN-like")
plt.legend()
```
